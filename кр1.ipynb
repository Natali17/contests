Среди строк, которые удовлетворяют условию, нужно запомнить первую и последнюю букву, и напечатать все эти буквы слитно в одну строку.

Примечание : кавычки в примерах даны для улучшения их читаемости, в вводе их не будет.

```python
prev = input()
arr = []
while True:
  if prev[0].istitle() == False:
    arr.append(prev[0])
    arr.append(prev[-1])
  curr = input()
  if curr == prev:
    break
  elif curr == '':
    break
  prev = curr
print(''.join(arr))
```

Пусть даны две последовательности ДНК: s = s 1 s 2 … s n и t = t 1 t 2 … t m . Будем говорить, что t может получится в результате эволюции из s, если s является подпоследовательностью t, то есть существует такая последовательность индексов 1 ≤ i 1 < i 2 < … < i n ≤ m, что s 1 =t i1 , s 2 =t i2 , … s n =t in . Необходимо выяснить, может ли последовательность t получится в результате эволюции из s.

```python
s = list(input())
t = list(input())

i = 0
for char in range(len(t)):
  if t[char] == s[i] and i < len(s):
    i += 1
  if i == len(s):
    break

if i == len(s):
  print('YES')
else:
  print('NO')
```

stop - остановиться
где N - натуральное число, расстояние на которое сдвинется робот в указанном в команде направлении.

Введём координатную плоскость, ось X направим вправо, ось Y вверх. Вам предоставляется информация о начальном положении робота, и последовательность команд, которые робот выполнит. Последовательность завершается командой остановки.

Необходимо определить, где окажется робот после выполнения набора команд.

```python
n, m = map(int, input().split())
while True:
  user_input = input() #dont write 'input = input()', to delete this, write 'del input'
  if user_input == 'stop':
    break
  action, coord = user_input.split(' ')
  if action == 'left':
    n -= int(coord)
  elif action == 'right':
    n += int(coord)
  elif action == 'down':
    m -= int(coord)
  elif action == 'top':
    m += int(coord)
print(n, m)
```

```python
n, m = map(int, input().split())
direction = []
numbers = []
while True:
    user_input = input()
    if user_input == 'stop':
        break
    direct, number = user_input.split()
    number = int(number)
    direction.append(direct)
    numbers.append(number)

for i in range(len(direction)):
    if direction[i] == 'left':
        n -= numbers[i]
    elif direction[i] == 'right':
        n += numbers[i]
    elif direction[i] == 'down':
        m -= numbers[i]
    elif direction[i] == 'top':
        m += numbers[i]
print(n, m)
```

#F
Градиентный спуск явлется одним из самых популярных методов оптимизации -- нахождения экстремума некоторого функционала.

Для работы метода необходимо знать функцию и ее первую производную. Метод заключается в следующем: мы берем некоторое начальное приближение экстремума. Затем итеративно вычисляем следующее приближение на основе предыдущего по формуле X n+1 = X n - a*f'(X n ), где a -- learning rate задает скорость, с которой сходится модель. Считается, что модель сошлась, когда изменение значения функции между соседними итерациями становится меньше или равным заданному порогу ε.

Ваша задача -- написать функцию def gradient_descent(func, derivative, x0, learning_rate, eps); , которая будет искать экстремум методом градиентного спуска. Функция должна вернуть пару (x i , f(x i )) с той итерации, на которой она сойдется.

Примечание: func и derivative являются объектами-функциями, которые можно использовать как обычные функции, т.е. func(x) .

image.png

```python
def gradient_descent(func, derivative, x0, learning_rate=0.1, eps=1e-4):
  x = x0
  while True:
    grad = derivative(x)
    x_new = x - learning_rate*grad
    if abs(func(x_new) - func(x)) <= eps:  # проверяем условие сходимости
      return x_new, func(x_new)
    x = x_new  # обновляем x для следующей итерации

result = gradient_descent(lambda x: x*x, lambda x: 2*x, 1)
print(result)
```

#G
Память в компьютере обычно разделена на ячейки. При записи данных в память, под них выделяется некоторое количество идущих подряд ячеек памяти. Программа при выполнении, так же, как и другие данные, размещается в оперативной памяти компьютера. Она представляет из себя набор инструкций. Так как инструкции могут выполняться не только последовательно, отдельно хранится информация о том, где в памяти находится следующая инструкция. Если программа могла бы обраться к ячейкам памяти, которые ей не принадлежат, она могла бы натворить много бед.

Однако операционная система всегда знает, где в памяти лежит программа и сколько ячеек она занимает. Поэтому, если программа пытается перейти к ячейкам памяти, которые ей не принадлежат, система может экстренно завершить программу и не дать ей испортить чужие данные. При этом возникает так называемая Ошибка Сегментации (Segmentation Fault).

Пусть информация о том, где в памяти хранится следующая инструкция задана смещением относительно текущей ячейки. Выполнение программы начинается с нулевой ячейки. Вам необходимо определить, из какой ячейки памяти происходит выход за ее границы.

Вам дается массив относительных адресов блоков памяти. Начав с 0-й позиции нужно переходить на ячейку, смещенную относительно текущей на число, указанное в текущем элемента массива. В конце необходимо вывести последний индекс перед выходом за границы массива.

image.png

```python
n = int(input())
arr = []
for i in range(n):
  arr.append(int(input()))

indexes = []
curr_position = 0
while 0 <= curr_position < len(arr):
    indexes.append(curr_position)
    curr_position += arr[curr_position]
print(indexes[-1])

#начать с нулевой позиции и перемещаться по массиву, каждый раз переходя на индекс,
#равный текущей позиции плюс значению в этой позиции. Нужно остановиться, когда
#следующий шаг выведет за границы массива, и вывести все индексы, на которые попал до выхода.
```

```python
n = int(input())
m = []
for i in range(n):
  i = int(input())
  m.append(i)

indexes = []
current_index = 0
while current_index < n:
  indexes.append(current_index)
  current_index += m[current_index]
print(indexes[-1])
```
