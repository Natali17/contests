```python
N = int(input())
count_1 = 0
count_2 = 0
count_3 = 0
count_4 = 0

for _ in range (N):
  i,j = map(int, input().split())
  if i > 0 and j > 0:
    count_1 += 1
  if i > 0 and j < 0:
    count_4 += 1
  if i < 0 and j > 0:
    count_2 += 1
  if i < 0 and j < 0:
    count_3 += 1
  if i == 0 or j == 0:
    continue

max_count = count_1
quarter = 1

if count_2 > max_count:
    max_count = count_2
    quarter = 2
if count_3 > max_count:
    max_count = count_3
    quarter = 3
if count_4 > max_count:
    max_count = count_4
    quarter = 4
print(quarter, max_count)




```

```python
N = int(input())
count_1 = 0
count_2 = 0
count_3 = 0
count_4 = 0

for _ in range(N):
    i, j = map(int, input().split())
    if i != 0 and j != 0:
      if  i > 0 and j > 0:
        count_1 += 1
      elif i > 0 and j < 0:
        count_4 += 1
      elif i < 0 and j < 0:
        count_3 += 1
      elif i < 0 and j > 0:
        count_2 += 1

max_count = max(count_1, count_2, count_3, count_4)
if max_count == count_1:
    print(1, max(count_1, count_2, count_3, count_4))
elif max_count == count_2:
    print(2, max(count_1, count_2, count_3, count_4))
elif max_count == count_3:
    print(3, max(count_1, count_2, count_3, count_4))
else:
    print(4, max(count_1, count_2, count_3, count_4))

```

```python
i = -1
array = []
while True:
  i = int(input())
  if i == 0:
    break
  array.append(i)

sum = 0
for i in range(0, len(array)):
  sum += array[i]
mean = sum/len(array)

square_sum = 0
deviations = 0
for i in range(0, len(array)):
  dev = (mean - array[i])
  deviations += dev**2
variance = deviations/len(array)
print(round(mean, 1), round(variance, 1))



```

```python
numbers = []
i = -1
while i != 0:
  i = int(input())
  numbers.append(i)
numbers.pop()

sqrmean_val = (sum(numbers)/len(numbers))**2
suma = 0
for i in numbers:
  suma += i**2
  mean_sqrval = suma/len(numbers)

mean_val = round(sum(numbers)/len(numbers), 3)
disp_val = round(mean_sqrval - sqrmean_val, 3)
print(mean_val, disp_val)
```
