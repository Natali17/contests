#B
Выяснить, является ли заданное число простым.

image.png

Для нахождения всех простых чисел не больше заданного числа n, следуя методу Эратосфена, нужно выполнить следующие шаги:

Выписать подряд все целые числа от двух до n (2, 3, 4, …, n).
Пусть переменная p изначально равна двум — первому простому числу.
Зачеркнуть в списке числа от 2p до n, считая шагами по p (это будут числа, кратные p: 2p, 3p, 4p, …).
Найти первое незачёркнутое число в списке, большее чем p, и присвоить значению переменной p это число.
Повторять шаги 3 и 4, пока возможно.
Теперь все незачёркнутые числа в списке — это все простые числа от 2 до n.

На практике, алгоритм можно улучшить следующим образом. На шаге № 3 числа можно зачеркивать, начиная сразу с числа p2, потому что все меньшие числа, кратные p, обязательно имеют простой делитель меньше p, и они уже будут зачеркнуты к этому времени. И, соответственно, останавливать алгоритм можно, когда p2 станет больше, чем n.[3] Кроме того, все простые числа, кроме 2, — нечётные числа, и поэтому для них можно считать шагами по 2p, начиная с p2.

```python
n = int(input())
array = list(range(n+1))
array[0] = array[1] = None

for i in range(2, int(n**0.5)):
  for j in range (i**2, n+1, i):
    array[j] = None

if array[n] == None:
  print(0)
else:
  print(1)

print(array)
```

```python
n = int(input())
l = list(range(n+1))
l[0]=l[2] = None
for i in range(2, n+1):
  for j in range(i * i, n + 1, i):
    l[j] = None
  if l[n] == None:
    break

if l[n] == None:
  print('0')
else:
  print('1')
```

 Всегда один делитель из пары меньше или равен sqrt(n), поэтому до этого числа можно делать проход, чтобы уменьшить врем работы цикла.

image.png

```python
n = int(input())
array = []
for i in range(2, n+1):
  while n % i == 0:
    n = n // i
    array.append(i)
  if n == 1:
    break
for i in array:
  print(i)

#Можно преобразовать числа в строки и объединить через '\n':
#print('\n'.join(map(str, array)))
```

Представьте натуральное число n в виде суммы двух кубов неотрицательных целых чисел. Т.е. n = a^3 + b^3, где n - натуральное, a и b - неотрицательные целые. Если такое представление возможно, выпишите найденные числа a и b через пробел в порядке неубывания. Если разложить нельзя, напишите строку "impossible".

image.png

```python
#for 'impossible' situations works way too long
n = int(input())
temp = 0
array = []
while temp != 1:
  for i in range(n+1):
    for j in range(n+1):
      if i**3 + j**3 == n:
        array.append(i)
        array.append(j)
        temp = 1
        break
    if temp == 1:
      break

if temp == 0:
  print('impossible')
else:
  print(*(array[0], array[1]))
```

```python
n = int(input())
found = False
for i in range(int(n**(1/3))+1):
  reminder = n - i**3
  j = round(reminder**(1/3))
  if i**3 + j**3 == n:
    print(i, j)
    found = True
    break

if found == False:
  print('impossible')
```

```python
number = int(input())
d = 0
k = 0
s = 0

while d**3 <= number / 2:
    c = number - d**3
    while k**3 <= c:
        if k**3 == c:
            print(d, k)
            s = 1
        k += 1
    if s == 1:
        break
    k = 0
    d += 1

if s == 0:
    print('impossible')
```

#G
Вам необходимо написать программу, вычисляющую минимальное число, для которого заданы: 1) сумма цифр, 2) количество цифр, 3) цифры должны быть упорядочены по неубыванию.

image.png

```python
number_sum, number_amount = map(int, input().split())

if number_sum > 9 * number_amount or number_sum < number_amount:
    print("impossible")
else:
    result = []
    for i in range(number_amount):
        # минимальная возможная цифра на этой позиции — та,
        # чтобы после неё оставалось достаточно суммы (минимум 1 на каждую оставшуюся позицию)
        min_digit = max(1, number_sum - 9 * (number_amount - i - 1)) #Какая минимальная цифра допустима на текущей позиции,
                                                                     #чтобы потом можно было «добрать» нужную сумму оставшимися цифрами
        digit = min(9, min_digit)
        result.append(str(digit))
        number_sum -= digit
    print(''.join(result))
```
