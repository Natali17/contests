Синтаксис по обработке строк <br>
Сравнение равенства строк происходит посимвольно:

>>> 'aaa' == 'aaa' <br>
>>> True <br>
>>> 'aab' == 'aaa' <br>
>>> False <br>

Проверка, что символ является десятичной цифрой:

>>> a, b = 'E', '8' <br>
>>> '0' <= a <= '9' <br>
>>> False <br>
>>> '0' <= b <= '9' <br>
>>> True <br>

Итерирование по символам строки в прямом порядке:

>>> s = 'abcd' <br>
>>> for c in s: <br>
...     print(c) <br>
... <br>
a <br>
b <br>
c <br>
d <br>

Итерирование по символам строки в обратном порядке:

>>> s = 'abcd' <br>
>>> for c in reversed(s): <br>
...     print(c) <br>
... <br>
d <br>
c <br>
b <br>
a <br>

Разбиение строки по подстроке:

>>> s = '912:adc:majw:d' <br>
>>> for part in s.split(':'): <br>
...     print(part) <br>
... <br>
912 <br>
adc <br>
majw <br>
d <br>

В данном примере разбиение происходит по символу : и строка разделяется на 4 части. По умолчанию строковый метод str.split() разбивает строку по символам пустого пространства (пробелы, табуляция).

#A
Для заданной строки символов вычислить произведение входящих в эту строку целых чисел (без учета их знаков).

image.png

````python
n = input()
multi = 1
temp = ''
for c in n:
  if '0' <= c <= '9':
    temp += c
  else:
    if temp != '':
      multi *= int(temp)
      temp = ''
if temp != '':
  multi *= int(temp)

print(multi)
````

#B
Главный врач хочет проанализировать статические данные потока поступающих пациентов. Он хочет увидеть среднюю арифметическую температуру пациентов от 60 лет и выше, а также пациентов, вес которых отклоняется от нормального больше, чем на 10 килограмм. Нормальным весом в килограммах считать рост за вычетом ста сантиметров. Если таких пациентов нет - вывести ноль.

image.png

````python
N = int(input())
data = []
for i in range(N):
    data.append(list(map(float, input().split()))) #data.append(map(float, input().split())) — это ошибка
temperature = 0
count = 0
for patient in data:
    if patient[0] >= 60 or abs(patient[2] - (patient[1] - 100)) > 10:
        temperature += patient[3]
        count += 1

if count == 0:
  print(0)
else:
  print(round(temperature/count, 5))
````

````python
N = int(input())
numbers = []
for i in range(N):
    numbers.append(list(map(float, input().split())))

suma = 0
count = 0
for i in range(N):
  norm_weight = numbers[i][1] - 100
  if numbers[i][0] >= 60 or abs(numbers[i][2] - norm_weight) > 10:
    suma += numbers[i][3]
    count += 1
if count > 0:
  mean_val = suma/count
else:
  mean_val = 0
print(round(mean_val, 5))
````

#C
Разложите неотрицательное целое десятичное число по степеням основания.

Слагаемое разложение должно быть записано слитно в формате: цифра*10^разряд . Слагаемые разделяются тремя символами: + (пробел, знак плюс, пробел). Слагаемые должны располагаться по убыванию разряда.

image.png

````python
n = int(input())
array = []
while n != 0:
  array.append(n%10)
  n = n // 10
array = array[::-1]

terms = []
for i in range(len(array)):
  digit = array[i]
  power = len(array) - i - 1
  term = f"{digit}*10^{power}"
  terms.append(term)
print(' + '.join(terms))
````

````python
n = int(input())
k = []
for i, digit in enumerate(str(n)):
  power = len(str(n)) - i - 1
  part = "{}*10^{}".format(digit, power) #вместо part = f"{digit}*10^{power}", тк версия Python ниже 3.6
  k.append(part)

print(" + ".join(k))
````

#D
Перевести число из него-десятичной системы счисления в десятичную.

image.png

````python
n = int(input())
array = []
for i, digit in enumerate(str(n)):
  digit = int(digit)
  power = len(str(n)) - i - 1
  value = digit * (-10)**power
  array.append(value)
suma = sum(array)
print(suma)
````

#E
Система счисления майя основывалась на 20-ричной системе счисления.

Разряд формируется следующим образом. Для записи нуля использовалась ракушка "@". Для конструирования числа разряда использовались точки ".", и палочки "|". Точка добавляла к числу разряда 1, а палочка 5. Например, число 13 помещается в один разряд и выглядит следующим образом: "...||" = 3 + 2*5.

Для отделения разрядов мы воспользуемся пробелом.

Вам предстоит считать число майя и перевести его в десятичную систему счисления.

image.png

````python
n = input()
k = []
temp = 0
for i in str(n):
  if i == '.':
    temp += 1
  elif i == '|':
    temp += 5
  elif i == '@':
    temp = 0
  elif i == ' ':
    k.append(temp)
    temp = 0
k.append(temp)

number = 0
for i, digit in enumerate(k):
  categ = len(k) - i - 1
  number += digit*20**categ
print(number)
````

#F
В вавилонской системе счисления для записи чисел внутри 60-ричного разряда использовались всего два знака: стоячий клин "v" (латинская строчная v) для обозначения единиц и лежачий клин "<" (знак строго меньше) для обозначения десятков. Для формирования разряда сначала выписывалось количество десятков, а затем количество единиц. Разряды вавилоняне разделяли пустым пространством, однако мы будем использовать для этого точку ".". Если какой-то разряд отсутствовал, на его месте ничего не писалось. Требуется перевести записанное в вавилонской системе число в десятичную систему счисления.


Ввод
v.v.<<v <br>
Вывод
3681

Ввод
vv..<<<vvvvv <br>
Вывод
7235

image.png

````python
n = input()
k = []
temp = 0
for i in str(n):
  if i == 'v':
    temp += 1
  elif i == '<':
    temp += 10
  elif i == '.':
    k.append(temp)
    temp = 0
k.append(temp)

number = 0
for i, digit in enumerate(k):
  categ = len(k) - i - 1
  number += digit*60**categ
print(number)
````
